Quick Sort

Quick Sort is a Divide and Conquer sorting algorithm.
It works by selecting a pivot element and arranging the array so that:
Elements smaller than pivot go to the left
Elements greater than pivot go to the right
Then the same process is repeated for the left and right parts.

Example
Sort the array:
64, 25, 32, 20, 40, 15
Choose the last element as pivot.

Step 1: First Partition
Pivot = 15
Compare all elements with 15.
64, 25, 32, 20, 40 | 15

After partition:
15, 25, 32, 20, 40, 64
Pivot 15 moves to its correct position.

Step 2: Next Partition
Now sort the remaining part:
25, 32, 20, 40, 64
Pivot = 64

All numbers are smaller than 64.
25, 32, 20, 40, 64
Step 3: Partition Again

Now sort:
25, 32, 20, 40
Pivot = 40

After partition:
25, 32, 20, 40
Step 4

Sort:
25, 32, 20
Pivot = 20

After partition:
20, 32, 25
Final Sorting
Continue recursively until sorted.

Final array:
15, 20, 25, 32, 40, 64

```python
def partition(array, left, right):

    pivot = array[left]
    i = left + 1
    j = right

    while True:

        while i <= j and array[i] <= pivot:
            i = i + 1

        while i <= j and array[j] > pivot:
            j = j - 1

        if i <= j:
            array[i], array[j] = array[j], array[i]
        else:
            break

    array[left], array[j] = array[j], array[left]

    return j


def quicksort(array, left, right):

    if left < right:

        pivot = partition(array, left, right)

        quicksort(array, left, pivot - 1)
        quicksort(array, pivot + 1, right)


# main program
arr = [23, 45, 12, 65, 34, 10, 3]

print("Original array:", arr)

quicksort(arr, 0, len(arr)-1)

print("Sorted array:", arr)
```
