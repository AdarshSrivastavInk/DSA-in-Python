1. Binary Tree
Definition

A Binary Tree is a tree where each node can have at most 2 children.

These children are called:

Left Child

Right Child

Example
        A
       / \
      B   C
     / \   \
    D   E   F
Explanation

A is the root node

B and C are children of A

D and E are children of B

👉 Each node has maximum two children, so it is a binary tree.

2. Binary Search Tree (BST)
Definition

A Binary Search Tree is a binary tree where:

Left child value < Parent node

Right child value > Parent node

Example
        50
       /  \
     30    70
    / \    / \
   20 40  60 80
Explanation

Numbers smaller than 50 go to the left

Numbers greater than 50 go to the right

👉 This makes searching faster.

3. Strict Binary Tree / Full Binary Tree
Definition

In a Strict Binary Tree, every node has:

0 children OR

2 children

Nodes cannot have only one child.

Example
        A
       / \
      B   C
     / \ / \
    D  E F  G
Explanation

A has 2 children

B has 2 children

C has 2 children

D,E,F,G have 0 children

👉 No node has only one child.

4. Complete Binary Tree
A Complete Binary Tree is a tree where:

All levels are completely filled

The last level fills from left to right

Example
        1
       / \
      2   3
     / \  /
    4  5 6
Explanation

Levels are filled from left side first

Last node appears on the left

👉 Used in Heap Data Structure.

5. Skewed Binary Tree
Definition

A Skewed Binary Tree is a tree where all nodes have only one child.

It looks like a linked list.

Types

1️⃣ Left Skewed Tree

      A
     /
    B
   /
  C
 /
D

2️⃣ Right Skewed Tree

A

     \
      D

👉 Tree becomes unbalanced.

6. Extended Binary Tree
Definition

An Extended Binary Tree is a binary tree where null children are replaced with special nodes.

These special nodes are called External Nodes.

Example

Normal Tree

      A
     / \
    B   C

Extended Tree

        A
       / \
      B   C
     / \ / \
    X  X X  X

Where X = external node (NULL)

Explanation

Every node now has exactly two children
Used in memory representation of trees

array represtion of tree


Array representation of a tree means storing the nodes of a binary tree inside an array instead of using pointers.
Each node is stored in an index position, and we use a simple formula to find its parent and children.
This method is commonly used in heap data structures.



Basic Rule of Array Representation
Assume the root starts at index 1.
For any node at index i:

| Relationship | Formula   |
| ------------ | --------- |
| Root         | index = 1 |
| Left Child   | 2 × i     |
| Right Child  | 2 × i + 1 |
| Parent       | i / 2     |


   Example Binary Tree
        A
       / \
      B   C
     / \   \
    D   E   F

 Array Representation
 | Index | Value |
| ----- | ----- |
| 1     | A     |
| 2     | B     |
| 3     | C     |
| 4     | D     |
| 5     | E     |
| 6     | NULL  |
| 7     | F     |



array from-
Index : 1  2  3  4  5  6  7

How Formula Works

Example: Node B

Index = 2

Left Child
2 × 2 = 4 → D
Right Child
2 × 2 + 1 = 5 → E

Parent
2 / 2 = 1 → A

```python
tree= [None]*10
def root(value):
  tree[1]= value
def left_child(parent_index,value):
  index = parent_index*2
  tree[index]= value
def right_child(parent_index,value):
  index = parent_index*2+1
  tree[index]= value
def print_tree():
  for i in range(1,9):
    print(tree[i],end=" ")

root("A")
left_child(1,"B")
right_child(1,"C")
left_child(2,"D")
right_child(2,"E")
right_child(3,"F")
print_tree()
```

linked list represtion

In linked list representation of a tree, each node is stored as a separate object and connected using links (pointers).

Each node contains three parts:

Data – value of the node

Left pointer – address of left child

Right pointer – address of right child

This method is also called linked representation of binary tree.

Structure of a Node
 -------------------------
|   Left   | Data | Right |
 -------------------------
Left → points to left child
Data → stores value
Right → points to right child

example binary tree
        A
       / \
      B   C
     / \   \
    D   E   F


A.left  → B
A.right → C

B.left  → D
B.right → E

C.right → F



```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Creating nodes
root = Node("A")
root.left = Node("B")
root.right = Node("C")

root.left.left = Node("D")
root.left.right = Node("E")

root.right.right = Node("F")

# Function to print tree (Preorder)
def preorder(node):
    if node:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)

preorder(root)
```
