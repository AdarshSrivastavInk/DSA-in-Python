```python
class node:
    def __init__(self, value):
        self.left = None
        self.right = None
        self.data = value

def insert(root, value):
    if root is None:
        return node(value)
    if root.data == value:
        return root
    if root.data > value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root

def search(root, value):
    if root is None:
        print("value not found")
        return
    if root.data == value:
        print("value found")
        return
    if root.data > value:
        search(root.left, value)
    else:
        search(root.right, value)

def delete(root,value):
  if(root==None):
    return root
  if (root.data>value):
    root.left=delete(root.left,value)
  if (root.data<value):
    root.right=delete(root.right,value)

  else:
    if(root.left == None):
      return root.right
    if(root.right == None):
      return root.left

def inorder(root):
    if root is not None:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)

# Building a proper BST
root = insert(None, 20)
root = insert(root, 10)
root = insert(root, 30)
root = insert(root, 40)
root = insert(root, 50)

print("Inorder Traversal:")
inorder(root)
print("\n")

search(root, 10)
search(root, 20)
search(root, 100)
delete(root,10)
print("Inorder Traversal:")
inorder(root)
print("\n")

delete(root,40)
print("Inorder Traversal:")
inorder(root)
print("\n")
```
