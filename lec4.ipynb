### Circular Linked List
In a circular linked list, the last node points back to the first node instead of `None`.
- **Singly Circular**: `last.next = head`
- **Traversal condition**: `while(t.next != self.head)`

```python
class node:
  def __init__(self,info,next=None):
    self.data = info
    self.next = next

class singlylinkedlist:
  def __init__(self,head = None):
    self.head = head

  def insertatend(self,value):
    new_node = node(value) # Create the new node
    if self.head is None: # If the list is empty
      self.head = new_node # The new node becomes the head
    else: # If the list is not empty
      current = self.head # Start from the head
      while current.next is not None: # Traverse to the last node
        current = current.next
      current.next = new_node # Append the new node

  def insertatbeg(self,value):
    new_node = node(value)
    new_node.next = self.head
    self.head = new_node

  def inserinmiddele(self,value,x): # Inserts 'value' after 'x'
    new_node = node(value)
    if self.head is None:
      print("List is empty, cannot insert in middle.")
      return

    current = self.head
    while current is not None:
      if current.data == x:
        new_node.next = current.next
        current.next = new_node
        return # Inserted, so exit
      current = current.next
    print(f"Value {x} not found in the list. Cannot insert {value}.")

  def deletell(self,value):
    current = self.head
    prev = None

    # Case 1: Head node itself holds the value to be deleted
    if current is not None and current.data == value:
      self.head = current.next
      current = None # Delete the node
      return

    # Search for the value to be deleted, keep track of the previous node
    while current is not None and current.data != value:
      prev = current
      current = current.next

    # If value was not present in linked list
    if current is None:
      print(f"Value {value} not found in the list.")
      return

    # Unlink the node from linked list
    prev.next = current.next
    current = None # Delete the node

  def printll(self):
    t1 = self.head
    if t1 is None:
      print("List is empty")
      return
    while(t1 is not None):
      print(t1.data, end=" -> ")
      t1 = t1.next
    print("None")

# Example usage:
obj = singlylinkedlist()
obj.insertatend(10)
obj.insertatend(20)
obj.insertatend(30)
obj.insertatend(40)
obj.insertatbeg(50)
obj.inserinmiddele(60,30) # Insert 60 after 30
obj.deletell(10)
obj.printll()
obj.deletell(70) # Test deleting non-existent value
obj.printll()
```

### Round-Robin and Linked Lists
Round-robin scheduling often uses a **Circular Linked List** because the last process naturally points back to the first one, allowing the scheduler to loop through processes indefinitely.

```python
class node:
  def __init__(self,info,next=None):
    self.data = info
    self.next = next

class singlylinkedlist:
  def __init__(self,head = None):
    self.head = head

  def insertatend(self,value):
    new_node = node(value)
    if self.head is None:
      self.head = new_node
    else:
      current = self.head
      while current.next is not None:
        current = current.next
      current.next = new_node

  def insertatbeg(self,value):
    new_node = node(value)
    new_node.next = self.head
    self.head = new_node

  def inserinmiddele(self,value,x):
    new_node = node(value)
    if self.head is None:
      return
    current = self.head
    while current is not None:
      if current.data == x:
        new_node.next = current.next
        current.next = new_node
        return
      current = current.next

  def deletell(self,value):
    current = self.head
    prev = None
    if current is not None and current.data == value:
      self.head = current.next
      return
    while current is not None and current.data != value:
      prev = current
      current = current.next
    if current:
      prev.next = current.next

  def printll(self):
    t1 = self.head
    while(t1 is not None):
      print(t1.data, end=' -> ')
      t1 = t1.next
    print('None')

obj = singlylinkedlist()
obj.insertatend(10)
obj.insertatend(20)
obj.insertatend(30)
obj.printll()
```

```python
class node:
  def __init__(self, info, next=None):
    self.data = info
    self.next = next

class circularlinkedlist:
  def __init__(self):
    self.head = None

  def insertatend(self, value):
    new_node = node(value)
    if self.head is None:
      self.head = new_node
      new_node.next = self.head
    else:
      current = self.head
      while current.next != self.head:
        current = current.next
      current.next = new_node
      new_node.next = self.head

  def printll(self):
    if self.head is None:
      print("List is empty")
      return
    current = self.head
    while True:
      print(current.data, end=" -> ")
      current = current.next
      if current == self.head:
        break
    print("(Back to Head)")

# Example usage for Round Robin simulation logic
cll = circularlinkedlist()
cll.insertatend("Process 1")
cll.insertatend("Process 2")
cll.insertatend("Process 3")
cll.printll()
```
