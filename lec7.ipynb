```python
class circularqueue:
  def __init__(self,size):
    self.size = size
    self.items = [None]*size
    self.front = -1
    self.rear = -1


  def enqueue(self,value):
    if (self.rear+1)%self.size == self.front:
      print("queue is full")
    elif self.front == -1:
      self.front = 0
      self.rear = 0
      self.items[self.rear] = value
    else:
      self.rear = (self.rear+1)%self.size
      self.items[self.rear] = value

  def dequeue(self):
    if self.front == -1:
      print("queue is empty")
    elif self.front == self.rear:
      print(self.items[self.front])
      self.front = -1
      self.rear = -1
    else:
      print(self.items[self.front])
      self.front = (self.front+1)%self.size

cq = circularqueue(5)
cq.enqueue(1)
cq.enqueue(2)
cq.enqueue(3)
cq.enqueue(4)
cq.enqueue(5)
cq.dequeue()
```
