Selection Sort is a sorting algorithm that repeatedly finds the smallest element from the unsorted part and places it at the beginning.

 In each pass:

Find the minimum element
Swap it with the first unsorted element

Key Idea
Selection Sort:
Find minimum element
Swap with first position
Repeat for remaining array


Example-
64, 25, 32, 20, 40, 15

Pass 1
Find the smallest element in the whole array.
Smallest = 15
Swap 64 and 15

15, 25, 32, 20, 40, 64


Pass 2
Find the smallest element from remaining list:
25, 32, 20, 40, 64
Smallest = 20
Swap 25 and 20
15, 20, 32, 25, 40, 64

Pass 3
Remaining list:
32, 25, 40, 64
Smallest = 25
Swap 32 and 25
15, 20, 25, 32, 40, 64

Pass 4
Remaining list:
32, 40, 64
Smallest = 32
No swap needed.

Pass 5
Remaining list:
40, 64
Already sorted.

Final Sorted Array
15, 20, 25, 32, 40, 64

```python
def SelectionSort(a):
  n = len(a)

  for i in range(n):
    min_idx = i
    for j in range(i + 1, n):
      if a[min_idx] > a[j]:
        min_idx = j
    a[i], a[min_idx] = a[min_idx], a[i]

# Test case moved outside the function
arr = [64, 32, 2545, 40, 15, 20]
SelectionSort(arr)
print(arr)
```
