# Bubble Sort

Bubble Sort is a simple sorting algorithm that repeatedly compares two adjacent elements and swaps them if they are in the wrong order.

 Larger elements “bubble up” to the end of the list, which is why it is called Bubble Sort.

25, 32, 20, 15, 45, 20, 15, 45, 64

final sorted areray

```python
def bubblesort(a):
  n = len(a)
  for i in range(n):
    for j in range(0,n-1-i):
      if a[j]>a[j+1]:
        a[j],a[j+1]=a[j+1],a[j]

a = [64,32,25,45,40,52,12]
bubblesort(a)
print(a)
```
