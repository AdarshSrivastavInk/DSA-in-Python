Breadth First Search (BFS) is a graph traversal algorithm used to visit all vertices of a graph.
 It explores nodes level by level.

In simple words:
Visit the starting node
Visit all its neighbors
Then visit neighbors of those nodes

```python
from collections import deque

class Graph:
    def __init__(self):
        self.adj = {}

    # add vertex
    def addvertex(self, v):
        if v not in self.adj:
            self.adj[v] = []

    # add edge
    def addedge(self, src, dest):
        self.addvertex(src)
        self.addvertex(dest)
        self.adj[src].append(dest)
        self.adj[dest].append(src)   # undirected graph

    # print graph
    def printgraph(self):
        for vertex in self.adj:
            print(vertex, "->", self.adj[vertex])

    # BFS traversal
    def bfs(self, src):
        visited = set()
        queue = deque([src])

        visited.add(src)

        while queue:
            vertex = queue.popleft()
            print(vertex, end=" ")

            for neighbor in self.adj[vertex]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)


# create graph
g = Graph()

g.addedge(0, 1)
g.addedge(0, 3)
g.addedge(1, 3)
g.addedge(3, 5)
g.addedge(3, 4)
g.addedge(4, 5)
g.addedge(4, 6)
g.addedge(6, 2)
g.addedge(6, 7)
g.addedge(7, 8)

g.printgraph()
print("\nBFS Traversal:")

g.bfs(0)
```
