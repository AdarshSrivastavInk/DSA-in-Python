```python
class stack:
  def __init__(self):
    self.s = []

  def length(self):
    return len(self.s)

  def push(self,value):
    self.s.insert(0,value)

  def peek(self):
    if len(self.s)==0:
      raise Exception("stack is empty")
    else:
      return self.s[0]

  def pop(self):
    if len(self.s)==0:
      raise Exception("stack is empty")
    else:
      return self.s.pop(0)

stk = stack()
stk.push(1)
stk.push(2)
stk.push(3)
stk.push(4)
print(stk.peek())
```
