Depth First Search (DFS) is a graph traversal algorithm used to visit all vertices of a graph.

 It explores as deep as possible first, then backtracks.
In simple words:
Go deep into the graph
When no more nodes are available, go back and explore another path

Algorithm Steps
Start at the starting vertex
Mark it visited
Visit one unvisited neighbor
Continue until no unvisited neighbor
Backtrack and repeat

Depth First Search (DFS) explores a graph as deep as possible first, then it comes back (backtracking) to explore other nodes.
To remember where to return, DFS needs a memory structure.
This memory is provided by a stack.

```python
class Graph:
    def __init__(self, size):
        self.size = size
        self.adj = {}   # dictionary to store adjacency list

    # function to add vertex
    def addvertex(self, v):
        if v not in self.adj:
            self.adj[v] = []

    # function to add edge
    def addedge(self, src, dest):
        self.addvertex(src)
        self.addvertex(dest)
        self.adj[src].append(dest)
        self.adj[dest].append(src)   # for undirected graph

    # function to print graph
    def printgraph(self):
        for vertex in self.adj:
            print(vertex, "->", self.adj[vertex])

    def dfs(self, src):
        # Using a set for visited nodes to handle arbitrary vertex names
        visited = set()
        stack = [src]

        while stack:
            v = stack.pop()
            if v not in visited:
                print(v, end=" ")
                visited.add(v)

                # Add neighbors to stack
                if v in self.adj:
                    for neighbor in reversed(self.adj[v]):
                        if neighbor not in visited:
                            stack.append(neighbor)

# create graph
g = Graph(6)

g.addedge(1, 2)
g.addedge(2, 3)
g.addedge(3, 4)
g.addedge(3, 5)
g.addedge(4, 5)
g.addedge(2, 4)

print("DFS traversal starting from vertex 1:")
g.dfs(1)
```
