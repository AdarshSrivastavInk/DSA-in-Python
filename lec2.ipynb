```python
from array import*

arr = array('i',[])

try:
    line = input("Enter numbers separated by space: ")
    # Split the input string by spaces and convert each part to an integer
    for x in line.split():
        arr.append(int(x))

    print("Array elements:")
    for x in arr:
      print(x, end=" ")
except ValueError:
    print("Invalid input! Please enter only integers separated by spaces.")
```
