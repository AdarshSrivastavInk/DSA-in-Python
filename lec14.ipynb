# Graph Adjency List Represtion


An Adjacency List is a way to represent a graph where each vertex stores a list of the vertices connected to it.

Instead of using a matrix, we keep a list of neighbors for each vertex.
This method uses less memory.

A ----- B
|       |
|       |
C ----- D

vertices-
A, B, C, D

Edges-
(A,B)
(A,C)
(B,D)
(C,D)

A connected to B and C
B connected to A and D
C connected to A and D
D connected to B and C


```python
class Graph:
    def __init__(self):
        self.adj = {}   # dictionary to store adjacency list

    # function to add vertex
    def addvertex(self, v):
        if v not in self.adj:
            self.adj[v] = []

    # function to add edge
    def addedge(self, src, dest):
        self.addvertex(src)
        self.addvertex(dest)
        self.adj[src].append(dest)
        self.adj[dest].append(src)   # for undirected graph

    # function to print graph
    def printgraph(self):
        for vertex in self.adj:
            print(vertex, "->", self.adj[vertex])


# create graph
g = Graph()

g.addedge(1,2)
g.addedge(2,3)
g.addedge(3,4)
g.addedge(3,5)
g.addedge(4,5)
g.addedge(2,4)

g.printgraph()
```
