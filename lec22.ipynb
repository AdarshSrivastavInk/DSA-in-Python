Divide and Conquer is an algorithm design technique where a big problem is:

 Divided into smaller subproblems
 Solved individually
 Combined to get the final solution

 In simple words:
Break the problem → Solve smaller parts → Combine results

Steps of Divide and Conquer
 Divide
Split the problem into smaller subproblems.

 Conquer
Solve the subproblems recursively.

 Combine
Combine the solutions of subproblems to solve the original problem.

```python
def find_min_max(arr, start, end):

    # if only one element
    if start == end:
        return arr[start], arr[start]

    # if two elements
    if start + 1 == end:
        if arr[start] < arr[end]:
            return arr[start], arr[end]
        else:
            return arr[end], arr[start]

    # divide array
    mid = (start + end) // 2

    min1, max1 = find_min_max(arr, start, mid)
    min2, max2 = find_min_max(arr, mid + 1, end)

    # combine results
    return min(min1, min2), max(max1, max2)


# main program
arr = [487,1789,4896,3685,8759]

minimum, maximum = find_min_max(arr, 0, len(arr)-1)

print("Minimum value is:", minimum)
print("Maximum value is:", maximum)
```
