An Undirected Graph is a graph where edges do not have any direction.

This means the connection between two vertices works both ways.

 If A is connected to B, then B is also connected to A.


A Directed Graph is a graph where edges have a direction.
This means the connection goes from one vertex to another in a specific direction.
 If A → B, it means A is connected to B, but B is not necessarily connected to A.
Edges are represented using arrows.

A Weighted Graph is a graph where each edge has a weight (value or cost).

The weight usually represents:
Distance
Cost
Time
Capacity
 In simple words, the connection between vertices has a number attached to it.

Graphs can be stored in two main ways:

Adjacency Matrix Representation
Definition

An Adjacency Matrix is a 2D array (matrix) used to represent a graph.

Rows represent source vertices

Columns represent destination vertices

1 = edge exists

0 = no edge

Matrix


|   | A | B | C | D |
| - | - | - | - | - |
| A | 0 | 1 | 1 | 0 |
| B | 1 | 0 | 0 | 1 |
| C | 1 | 0 | 0 | 1 |
| D | 0 | 1 | 1 | 0 |


Explanation

A connected to B → 1

A connected to C → 1

A not connected to D → 0

Adjacency List Representation
Definition

An Adjacency List represents a graph using a list of connected vertices.

Each vertex stores its neighbors.

Representation
A → B, C
B → A, D
C → A, D
D → B, C

```python
class Graph:
  def __init__(self,vertax):
    self.mat = [[0]*vertax for x in range(vertax)]
    self.size = vertax

  def add_edge(self,src,dest):
    if(0<=src<self.size and 0<=dest<self.size):
      self.mat[src][dest]=1
      self.mat[dest][src]=1 #undirected graph

    else:
      print("invalid edge")

  def print(self):
    for row in self.mat:
      print(" ".join(map(str,row)))

g = Graph(7)
g.add_edge(0,1)
g.add_edge(0,2)
g.add_edge(0,3)
g.add_edge(1,4)
g.add_edge(2,5)
g.add_edge(3,6)
g.add_edge(4,5)
g.add_edge(5,6)
g.add_edge(1,3)
g.add_edge(2,4)
g.print()
```
