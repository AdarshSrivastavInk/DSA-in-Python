A Binary Search Tree (BST) is a special type of binary tree where nodes are arranged in a specific order.

Rules of BST:

Left child value is smaller than the parent

Right child value is greater than the parent

Both left and right subtrees also follow the same rule.

👉 This ordering helps search data quickly.

 Example of Binary Search Tree


         50
       /  \
     30    70
    / \    / \
   20 40  60 80




   Explanation

For node 50

Left side numbers → 30, 20, 40 (smaller)

Right side numbers → 70, 60, 80 (greater)

For node 30

Left → 20

Right → 40

The same rule applies to every node.

```python
class node:
    def __init__(self, value):
        self.left = None
        self.right = None
        self.data = value

def insert(root, value):
    if root is None:
        return node(value)
    if root.data == value:
        return root
    if root.data > value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root

def inorder(root):
    if root is not None:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)

# Initialize and build the tree
root = node(20)
insert(root, 10)
insert(root, 30)
insert(root, 5)
insert(root, 15)

print("Inorder Traversal:")
inorder(root)
```

```python
class node:
    def __init__(self, value):
        self.left = None
        self.right = None
        self.data = value

def insert(root, value):
    if root is None:
        return node(value)
    if root.data == value:
        return root
    if root.data > value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root

def search(root, value):
    if root is None:
        print("value not found")
        return
    if root.data == value:
        print("value found")
        return
    if root.data > value:
        search(root.left, value)
    else:
        search(root.right, value)

def inorder(root):
    if root is not None:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)

# Building a proper BST
root = insert(None, 20)
root = insert(root, 10)
root = insert(root, 30)
root = insert(root, 40)
root = insert(root, 50)

print("Inorder Traversal:")
inorder(root)
print("\n")

search(root, 10)
search(root, 20)
search(root, 100)
```
