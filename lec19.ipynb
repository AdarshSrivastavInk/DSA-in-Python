Insertion Sort is a sorting algorithm that builds the sorted list one element at a time.

 It works like arranging playing cards in your hand.
You pick one element and insert it into the correct position in the sorted part.

Example
Sort the array:
64, 25, 32, 20, 40, 15


Pass 1
Take 25 and compare with 64.
64, 25, 32, 20, 40, 15
↓
25, 64, 32, 20, 40, 15
Sorted part:
25, 64


Pass 2
Take 32 and insert into correct position.
25, 64, 32, 20, 40, 15
↓
25, 32, 64, 20, 40, 15
Sorted part:
25, 32, 64


Pass 3
Insert 20.
25, 32, 64, 20, 40, 15
↓
20, 25, 32, 64, 40, 15
Sorted part:
20, 25, 32, 64


Pass 4
Insert 40.
20, 25, 32, 64, 40, 15
↓
20, 25, 32, 40, 64, 15


Pass 5
Insert 15.
20, 25, 32, 40, 64, 15
↓
15, 20, 25, 32, 40, 64

Final Sorted Array
15, 20, 25, 32, 40, 64

```python
def insertionsort(a):
    n = len(a)

    for i in range(1, n):
        key = a[i]
        j = i - 1

        while j >= 0 and key < a[j]:
            a[j+1] = a[j]
            j = j - 1

        a[j+1] = key


# main program
a = [64, 25, 32, 20, 40, 15]

print("Original array:", a)

insertionsort(a)

print("Sorted array:", a)
```
