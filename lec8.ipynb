factorial number is n!
n! = n-1! *n!

# factorial(n) = n * factorial(n-1) -> recursion
def factorial(n):
    # Base case: if n is 0 or 1, the factorial is 1
    if n == 0 or n == 1:
        return 1
    # Recursive step: n * factorial of (n-1)
    else:
        return n * factorial(n - 1)

# Test the function
number = 5
print(f"The factorial of {number} is {factorial(number)}")

fibonnaci series -> 1,2,3,5,8,13,21

```python
def fib(n):
  if n ==1 or n==2:
    return 1
  else:
    return fib(n-1)+fib(n-2)

num =7
print(f"the fibnocci series {num} is {fib(num)}")
```
