Tree Traversal means visiting every node of a tree exactly once in a specific order.
In simple words:
👉 It is the process of reading or accessing all nodes of a tree.
There are three main types of tree traversal in a binary tree.

Types of Tree Traversal->


Preorder Traversal
Inorder Traversal
Postorder Traversal


example
        A
       / \
      B   C
     / \   \
    D   E   F


    

Preorder Traversal

Root → Left → Right

Steps->

Visit Root
Traverse Left subtree
Traverse Right subtree

example->
A B D E C F
Explanation:
A → B → D → E → C → F

 Inorder Traversal

 Left → Root → Right

 Steps->
Traverse Left subtree
Visit Root
Traverse Right subtree

example-

 Postorder Traversal
 Left → Right → Root

 Steps->

Visit Root

```python
class node:
    def __init__(self,value):
      self.left = None
      self.right = None
      self.data = value

def preorder(root):
    if root is not None:
        print(root.data, end=" ")
        preorder(root.left)
        preorder(root.right)


root = node(1)
root.left = node(2)
root.right = node(3)
root.left.left = node(4)
root.left.right = node(5)
root.right.right = node(6)

print("Preorder Traversal:")
preorder(root)
```

```python
class node:
    def __init__(self,value):
      self.left = None
      self.right = None
      self.data = value

def preorder(root):
    if root is not None:
        print(root.data, end=" ")
        preorder(root.left)
        preorder(root.right)

def inorder(root):
  if(root !=None):
    inorder(root.left)
    print(root.data,end=" ")
    inorder(root.right)

def postorder(root):
  if(root !=None):
    postorder(root.left)
    postorder(root.right)
    print(root.data,end=" ")

root = node(1)
root.left = node(2)
root.right = node(3)
root.left.left = node(4)
root.left.right = node(5)
root.right.right = node(6)

print("Preorder Traversal:")
preorder(root)
print('\n')
print("inorder Travesal:")
inorder(root)
print('\n')
print("postorder Travesal:")
postorder(root)
```
