Merge Sort

Merge Sort is a Divide and Conquer sorting algorithm.

It works in two main steps:

Divide the array into smaller parts
 Merge the sorted parts back together

 The array keeps splitting until each part has one element, then they are merged in sorted order.

Example-
Sort the array:
64, 25, 32, 20, 40, 15


Step 1: Divide the Array
Split into two parts:
64, 25, 32   |   20, 40, 15
Split again:
64 | 25, 32      20 | 40, 15
Split again:
64 | 25 | 32      20 | 40 | 15
Now each part has one element.

Step 2: Merge Sorted Arrays
Merge small parts:
25, 32
64 + (25,32) → 25, 32, 64
Other side:
40, 15 → 15, 40
20 + (15,40) → 15, 20, 40

Final Merge
25, 32, 64   +   15, 20, 40
Final sorted array:
15, 20, 25, 32, 40, 64

divide (left,right)

first(left to m-1)
second(m to right)


middle = left+right/2

```python
def merge(arr, left, mid, right):

    left_part = arr[left:mid+1]
    right_part = arr[mid+1:right+1]

    i = 0
    j = 0
    k = left

    while i < len(left_part) and j < len(right_part):
        if left_part[i] <= right_part[j]:
            arr[k] = left_part[i]
            i += 1
        else:
            arr[k] = right_part[j]
            j += 1
        k += 1

    while i < len(left_part):
        arr[k] = left_part[i]
        i += 1
        k += 1

    while j < len(right_part):
        arr[k] = right_part[j]
        j += 1
        k += 1


def divide(arr, left, right):

    if left < right:

        mid = (left + right) // 2

        divide(arr, left, mid)
        divide(arr, mid + 1, right)

        merge(arr, left, mid, right)


# Main Program
arr = [64, 25, 32, 20, 40, 15]

print("Original Array:", arr)

divide(arr, 0, len(arr)-1)

print("Sorted Array:", arr)
```
